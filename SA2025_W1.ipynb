### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
#Multi-line plot: average mpg over years per origin
avg_mpg_by_year_origin = df.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()
plt.figure(figsize=(12, 6))
sns.lineplot(data=avg_mpg_by_year_origin, x='model_year', y='mpg', hue='origin', marker='o')
plt.title('Average MPG Over Years by Origin')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### We define a “consistent” car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
## Group by car name and count unique model_years
car_year_counts = df.groupby('name')['model_year'].nunique()

# Filter names that appear in more than one model year
repeated_models = car_year_counts[car_year_counts > 1].index

# Filter original dataframe for these repeated models
repeated_df = df[df['name'].isin(repeated_models)]

# Group by name and compute standard deviation, count, and mean of mpg
stats = repeated_df.groupby('name')['mpg'].agg(['std', 'count', 'mean']).reset_index()

# Filter for std deviation < 1.0 and count >= 2
consistent_models = stats[(stats['std'] < 1.0) & (stats['count'] >= 2)]

# Sort by count (descending) then mean mpg (descending)
consistent_models_sorted = consistent_models.sort_values(by=['count', 'mean'], ascending=[False, False])

# Rename columns for clarity
consistent_models_sorted.rename(columns={
    'name': 'Model Name',
    'count': 'Number of Appearances',
    'mean': 'Average MPG',
    'std': 'MPG Std Dev'
}, inplace=True)

# Show top consistent models
consistent_models_sorted.reset_index(drop=True, inplace=True)
consistent_models_sorted.head(10)  # Display top 10 consistent models
```
